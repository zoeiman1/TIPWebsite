#Loading Dataset
The load_data function is designed to load the internet traffic dataset. It attempts to load from a provided URL first. If that fails, it tries to load from a local file. If both fail, it resorts to a hardcoded fallback URL to ensure the data is loaded. The loaded data is then stored in a pandas DataFrame df.

```python
def load_data(data_url=None):
    if data_url:
        try:
            print(f"\n📥 Loading from URL: {data_url}")
            df = pd.read_csv(data_url)
            print(f"✓ Loaded {len(df)} records from URL")
            return df
        except Exception as e:
            print(f"⚠ Could not load from URL: {e}")
            # If provided URL fails, try local file
            try:
                print("\n📁 Loading local file: cs448b_ipasn.csv")
                df = pd.read_csv('cs448b_ipasn.csv')
                print(f"✓ Loaded {len(df)} records")
                return df
            except Exception as e_local:
                print(f"⚠ Could not load local file: {e_local}")
                # If local file also fails, try the hardcoded URL
                # This recreates the intent of the original DATA_URL and recursive call
                FALLBACK_URL = "https://raw.githubusercontent.com/Fishstickontop/Internet-Traffic-Prediction/refs/heads/main/cs448b_ipasn.csv"
                print(f"\nAttempting to load from fallback URL via recursive call: {FALLBACK_URL}")
                return load_data(FALLBACK_URL) # Recursive call with the hardcoded URL
    else: # If no data_url was provided initially
        # Try local file first
        try:
            print("\n📁 Loading local file: cs448b_ipasn.csv")
            df = pd.read_csv('cs448b_ipasn.csv')
            print(f"✓ Loaded {len(df)} records")
            return df
        except Exception as e_local:
            print(f"⚠ Could not load local file: {e_local}")
            # If local file fails, try the hardcoded URL
            FALLBACK_URL = "https://raw.githubusercontent.com/Fishstickontop/Internet-Traffic-Prediction/refs/heads/main/cs448b_ipasn.csv"
            print(f"\nAttempting to load from fallback URL via recursive call: {FALLBACK_URL}")
            return load_data(FALLBACK_URL) # Recursive call with the hardcoded URL

# The original call to load_data should now be outside the function definition.
# For example, like this:
DATA_URL = "https://raw.githubusercontent.com/Fishstickontop/Internet-Traffic-Prediction/refs/heads/main/cs448b_ipasn.csv"
df = load_data(DATA_URL)
```

#8. Creating Time Series Datasets
The create_dataset function transforms our scaled time series data into a supervised learning problem. It creates input-output pairs where X consists of sequences of look_back time steps, and y consists of the subsequent look_back time steps as targets. This is a common practice for sequence prediction tasks like time series forecasting. The shapes of the resulting X_train, y_train, X_test, and y_test are printed to verify the data structure.

```python
def create_dataset(dataset, look_back):
    X , y = [], []
    for i in range(len(dataset)-look_back):
        feature = dataset[i:i+look_back]
        target = dataset[i+1:i+look_back+1]
        X.append(feature)
        y.append(target)
    return np.array(X), np.array(y)
```

#Quantum Convolutional Layers
 This layer is designed to integrate quantum computing principles into a convolutional neural network (CNN) architecture. It uses cirq to build quantum circuits and tensorflow_quantum to simulate or execute them within a TensorFlow model.

Here's a breakdown of its key components:

__init__: Initializes the layer, setting up parameters like filter_size (the size of the quantum filter, which must be a power of 2), depth (number of quantum convolution layers), and an activation function.

_get_new_param: Generates new symbolic parameters for the quantum gates, which will become trainable weights in the Keras model.

_QConv: Defines the quantum convolutional operation, which consists of learnable CZPowGate and CXPowGate applications between qubits.

QCNN_layer_gen: Constructs the quantum circuit for a single quantum convolutional filter. It first maps classical input data onto qubits using cirq.rx gates and then applies multiple layers of the learnable _QConv gates to create the convolutional effect.

build: This method is called when the layer is first used and helps in defining the input shape and initializing kernel weights. It calculates the number of times the quantum filter will slide across the input data (num_x, num_y).

call: This is the main method where the forward pass of the layer occurs. It takes the input tensor, slices it into overlapping patches (similar to a classical CNN), flattens these patches, and then feeds each patch through the quantum circuit defined in QCNN_layer_gen. The quantum circuit's output (expectation values) is then reshaped and passed through the specified activation function.

single_depth_QCNN: Processes the input data for a single 'depth' or channel, combining the input classical values with the trainable quantum parameters and calculating the expectation value of an observable (cirq.Z) on the quantum circuit.



```python
class QConv(tf.keras.layers.Layer):
    def __init__(self, filter_size, depth, activation=None, name=None, kernel_regularizer=None, **kwangs):
        super(QConv, self).__init__(name=name, **kwangs)
        self.filter_size = filter_size
        self.depth = depth
        self.learning_params = []
        self.QCNN_layer_gen()
        # self.circuit_tensor = tfq.convert_to_tensor([self.circuit])
        self.activation = tf.keras.layers.Activation(activation)
        self.kernel_regularizer = kernel_regularizer

    def _next_qubit_set(self, original_size, next_size, qubits):
        step = original_size // next_size
        qubit_list = []
        for i in range(0, original_size, step):
            for j in range(0, original_size, step):
                qubit_list.append(qubits[original_size*i + j])
        return qubit_list

    def _get_new_param(self):
        """
        return new learnable parameter
        all returned parameter saved in self.learning_params
        """
        new_param = sympy.symbols("p"+str(len(self.learning_params)))
        self.learning_params.append(new_param)
        return new_param

    def _QConv(self, step, target, qubits):
        """
        apply learnable gates each quantum convolutional layer level
        """
        yield cirq.CZPowGate(exponent=self._get_new_param())(qubits[target], qubits[target+step])
        yield cirq.CXPowGate(exponent=self._get_new_param())(qubits[target], qubits[target+step])

    def QCNN_layer_gen(self):
        """
        make quantum convolutional layer in QConv layer
        """
        pixels = self.filter_size**2
        # filter size: 2^n only for this version!
        if np.log2(pixels) % 1 != 0:
            raise NotImplementedError("filter size: 2^n only available")
        cirq_qubits = cirq.GridQubit.rect(self.filter_size, self.filter_size)
        # mapping input data to circuit
        input_circuit = cirq.Circuit()
        input_params = [sympy.symbols('a%d' %i) for i in range(pixels)]
        for i, qubit in enumerate(cirq_qubits):
            input_circuit.append(cirq.rx(np.pi*input_params[i])(qubit))
        # apply learnable gate set to QCNN circuit
        QCNN_circuit = cirq.Circuit()
        step_size = [2**i for i in range(np.log2(pixels).astype(np.int32))]
        for step in step_size:
            for target in range(0, pixels, 2*step):
                QCNN_circuit.append(self._QConv(step, target, cirq_qubits))
        # merge the circuits
        full_circuit = cirq.Circuit()
        full_circuit.append(input_circuit)
        full_circuit.append(QCNN_circuit)
        self.circuit = full_circuit # save circuit to the QCNN layer obj.
        self.params = input_params + self.learning_params
        self.op = cirq.Z(cirq_qubits[0])

    def build(self, input_shape):
        self.width = input_shape[1]
        self.height = input_shape[2]
        self.channel = input_shape[3]
        self.num_x = self.width - self.filter_size + 1
        self.num_y = self.height - self.filter_size + 1

        self.kernel = self.add_weight(name="kenel",
                                      shape=[self.depth,
                                             self.channel,
                                             len(self.learning_params)],
                                     initializer=tf.keras.initializers.glorot_normal(),
                                     regularizer=self.kernel_regularizer)
        self.circuit_tensor = tfq.convert_to_tensor([self.circuit] * self.num_x * self.num_y * self.channel)

    def call(self, inputs):
        # input shape: [N, width, height, channel]
        # slide and collect data
        stack_set = None
        for i in range(self.num_x):
            for j in range(self.num_y):
                slice_part = tf.slice(inputs, [0, i, j, 0], [-1, self.filter_size, self.filter_size, -1])
                slice_part = tf.reshape(slice_part, shape=[-1, 1, self.filter_size, self.filter_size, self.channel])
                if stack_set == None:
                    stack_set = slice_part
                else:
                    stack_set = tf.concat([stack_set, slice_part], 1)
        # -> shape: [N, num_x*num_y, filter_size, filter_size, channel]
        stack_set = tf.transpose(stack_set, perm=[0, 1, 4, 2, 3])
        # -> shape: [N, num_x*num_y, channel, filter_size, fiter_size]
        stack_set = tf.reshape(stack_set, shape=[-1, self.filter_size**2])
        # -> shape: [N*num_x*num_y*channel, filter_size^2]

        # total input citcuits: N * num_x * num_y * channel
        circuit_inputs = tf.tile([self.circuit_tensor], [tf.shape(inputs)[0], 1])
        circuit_inputs = tf.reshape(circuit_inputs, shape=[-1])
        tf.fill([tf.shape(inputs)[0]*self.num_x*self.num_y, 1], 1)
        outputs = []
        for i in range(self.depth):
            controller = tf.tile(self.kernel[i], [tf.shape(inputs)[0]*self.num_x*self.num_y, 1])
            outputs.append(self.single_depth_QCNN(stack_set, controller, circuit_inputs))
            # shape: [N, num_x, num_y]

        output_tensor = tf.stack(outputs, axis=3)
        output_tensor = tf.math.acos(tf.clip_by_value(output_tensor, -1+1e-5, 1-1e-5)) / np.pi
        # output_tensor = tf.clip_by_value(tf.math.acos(output_tensor)/np.pi, -1, 1)
        return self.activation(output_tensor)

    def single_depth_QCNN(self, input_data, controller, circuit_inputs):
        """
        make QCNN for 1 channel only
        """
        # input shape: [N*num_x*num_y*channel, filter_size^2]
        # controller shape: [N*num_x*num_y*channel, len(learning_params)]
        input_data = tf.concat([input_data, controller], 1)
        # input_data shape: [N*num_x*num_y*channel, len(learning_params)]
        QCNN_output = tfq.layers.Expectation()(circuit_inputs,
                                               symbol_names=self.params,
                                               symbol_values=input_data,
                                               operators=self.op)
        # QCNN_output shape: [N*num_x*num_y*channel]
        QCNN_output = tf.reshape(QCNN_output, shape=[-1, self.num_x, self.num_y, self.channel])
        return tf.math.reduce_sum(QCNN_output, 3)
```

* Model Initialization: It creates a Sequential



Keras model.
* Quantum Convolutional Layer: It adds a custom QConv layer as the first layer. This layer takes parameters like filter_size=1, depth=8 (indicating 8 quantum convolution layers), and uses a linear activation function. The input_shape is dynamically set based on the dimensions of your training data (X_train).
* Classical Layers: After the quantum convolutional layer, the output is flattened and then passed through two standard dense (fully connected) layers with relu activation. This helps in extracting higher-level features from the quantum layer's output.
* Output Layer: The final layer is another Dense layer with look_back units and a linear activation. This is crucial for time series forecasting, as it predicts look_back future values based on the input sequence.
* Compilation: The model is compiled using the SGD (Stochastic Gradient Descent) optimizer, mean_squared_error as the loss function (common for regression tasks), and mae (Mean Absolute Error) as a metric to evaluate performance.
* Training: The model is trained using the model.fit method on X_train and y_train for 50 epochs, with a batch_size of 5. It also evaluates performance on the validation set (X_test, y_test) during training.
* Prediction: After training, the model generates predictions (y_pred) on the X_test dataset.
Return Values: The function returns the trained model, its training history (which includes loss and metrics per epoch), and the predictions made on the test se

```python
def trainQCNN(QCNN):
  width = np.shape(X_train)[1]
  height = np.shape(y_train)[2]
  model = models.Sequential()
  model.add(QConv(filter_size=1, depth=8, activation='linear',
                      name='qconv1', input_shape=(width, height, 1)))
  #model.add(layers.Reshape((width, 8)))
  model.add(layers.Flatten())
  model.add(layers.Dense(units = 64, activation='relu'))
  model.add(layers.Dense(units = 32, activation='relu'))
  model.add(layers.Dense(look_back, activation='linear')) # Change output layer to match target shape
  model.compile(optimizer=tf.optimizers.SGD(),
              loss='mean_squared_error',
              metrics=['mae'])
  history = model.fit(X_train, y_train, validation_data=(X_test, y_test),  epochs=50, batch_size=5)
  y_pred = model.predict(X_test)
  print(y_pred)
  return model, history, y_pred
```
